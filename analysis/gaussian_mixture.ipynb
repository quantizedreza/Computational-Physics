```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the Gaussian Mixture
N = 2
mu1, mu2 = -1, 1
sigma1, sigma2 = 0.1, 0.1
w1, w2 = 0.5, 0.5

# Generate x values
x = np.linspace(-3, 3, 1000)

# Calculate the PDF for each Gaussian
y1 = w1 * norm.pdf(x, mu1, sigma1)
y2 = w2 * norm.pdf(x, mu2, sigma2)

# Combine the PDFs
y_mixture = y1 + y2

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label=f'Gaussian 1: ={mu1}, ={sigma1}', linestyle='--')
plt.plot(x, y2, label=f'Gaussian 2: ={mu2}, ={sigma2}', linestyle='--')
plt.plot(x, y_mixture, label='Gaussian Mixture', linewidth=2)

plt.title('PDF of Gaussian Mixture Model')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the Gaussian Mixture
N = 2
mu1, mu2 = -1, 1
sigma1, sigma2 = 1, 1  # Changed to 1
w1, w2 = 0.5, 0.5

# Generate x values
x = np.linspace(-4, 4, 1000)

# Calculate the PDF for each Gaussian
y1 = w1 * norm.pdf(x, mu1, sigma1)
y2 = w2 * norm.pdf(x, mu2, sigma2)

# Combine the PDFs
y_mixture = y1 + y2

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label=f'Gaussian 1: ={mu1}, ={sigma1}', linestyle='--', color='blue')
plt.plot(x, y2, label=f'Gaussian 2: ={mu2}, ={sigma2}', linestyle='--', color='red')
plt.plot(x, y_mixture, label='Gaussian Mixture', linewidth=2, color='blue')

plt.title('The pdf of Gaussian Mixture with =1')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()

# Save the figure
plt.savefig('gaussian_mixture_pdf_sigma1.png', dpi=300, bbox_inches='tight')

# Show the plot (optional, if you want to see it)
plt.show()
```
